```python
# find the min, max, mean and std of the Heart column

min = df['Heart'].min()
max = df['Heart'].max()
mean = df['Heart'].mean()
std = df['Heart'].std()

print("Min, Max, Mean, Std : ", min, max, mean, std)
print('--------------------------------------------')
print(f"Min: {min}, Max: {max}, Mean: {mean}, Std: {std}")
```

```python
# noramlize the Heart column using z-scoring

(df['Heart']- mean) / std
```

```python
# noramlize the Heart column using minmax scaling

(df['Heart']- min) / (max - min)
```

```python
# are there any non-number (missing) values in this column?

df['Heart'].isna()
```

```python
# Fill NaN Values in the Heart Column with the Mean value (impute)

df['Heart'] = df['Heart'].fillna(df['Heart'].mean())
```

```python
# Remove negative values for Alcohol

df['Alcohol'] = df['Alcohol'].abs()
```

```python
# Data reduction for Heart

df['Heart'].round(decimals = 1)
```

```python
# examine the partially cleansed data

df
```

```python
# Let's add the wrangled data for India

df = df.append({'Country': 'India', 'Alcohol':2.53, 'Deaths': 'NA', 'Heart': 171, 'Liver': 'NA'}, ignore_index=True)

```

```python

"""  Discretize Alcohol

o	0: less than or equal to 1.00 per capita
o	1: more than 1.00 but less than or equal to 2.00 per capita
o	2: more than 2.00 but less than or equal to 5.00 per capita
o	3: more than 5.00 per capita

"""

df['Alcohol'] = pd.cut(x=df['Alcohol'], bins=[0, 1, 2, 5, 100], labels = [0, 1, 2, 3])

df
```

```python
#Change "NA" to NaN
df['Deaths'] = pd.to_numeric(df['Deaths'], errors='coerce')
```

```python
#Fill in NaN with mean of column
df['Deaths'] = df['Deaths'].fillna(df['Deaths'].mean())
```

```python
#Change all negative numbers to positive
df['Deaths'] = df['Deaths'].abs()
```

```python
#Change "NA" to NaN
df['Liver'] = pd.to_numeric(df['Liver'], errors='coerce')
```

```python
#Fill in NaN with mean of column
df['Liver'] = df['Liver'].fillna(df['Liver'].mean())
```

```python
"""
  Discretize Deaths:
o	6: greater than or equal to 600 but less than 700 per captita
o	7: greater than or equal to 700 but less than 800 per captita
o	8: greater than or equal to 800 but less than 900 per captita
o	9: greater than or equal to 900 but less than 1000 per captita
o 10: greater than or equal to 1000 per capita


"""

df['Deaths'] = pd.cut(x=df['Deaths'], bins=[600, 700, 800, 900, 1000, 100000], labels = [6, 7, 8, 9, 10])
df
```

```python
#delete India row
df = df.drop(21)
```

```python
#delete Alcohol column
del df['Alcohol']
df
```

```python
df['Heart'] = df['Heart'].round(decimals = 1)
df['Liver'] = df['Liver'].round(decimals = 1)
df
```
